Player Data Cleaning

```python
players = pd.read_csv("players.csv")
```

```python
players
```

```python
del players["Unnamed: 0"]
del players["Rk"]
```

```python
players["Player"].head(30)
```

```python
players["Player"] = players["Player"].str.replace("*", "", regex=False)
```

```python
players
```

```python
players.head(30)
```

```python
players["Player"].head(30)
```

```python
def single_data(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row
    
players = players.groupby(["Player", "Year"]).apply(single_data)
```

```python
players.head(20)
```

```python
players.index = players.index.droplevel()
```

```python
nicknames = {}

with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev,name =line.replace("\n","").split(",")
        nicknames[abbrev] = name
```

```python
merged["Team"] = merged["Tm"].map(nicknames)
```
