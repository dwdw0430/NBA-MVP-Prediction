```python
import pandas as pd
```

```python
stats = pd.read_csv("player_advanced_stats.csv")
```

```python
del stats["Unnamed: 0"]
```

```python
pd.isnull(stats).sum()
```

```python
stats = stats.fillna(0)
```

```python
pd.isnull(stats).sum()
```

```python
stats.columns
```

```python
stats_2023 = pd.read_csv("player_team_2023.csv")
```

```python
del stats_2023["Unnamed: 0"]
```

```python
pd.isnull(stats_2023).sum()
```

```python
stats_2023 = stats_2023.fillna(0)
```

```python
pd.isnull(stats_2023).sum()
```

```python
predictors = ['Age', 'G', 'GS', 'MP', 'FG', 'FGA', 'FG%', '3P',
       '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB',
       'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'Year', 'W', 
       'L', 'W/L%', 'GB', 'PS/G','PA/G', 'SRS', 'PER', 'TS%', 'USG%', 'WS', 'WS/48',
       'BPM', 'VORP']
```

```python
stats_2023["Year"]
```

```python
train = stats[stats["Year"] < 2023]
```

```python
test = stats_2023[stats_2023["Year"] == 2023]
```

```python
from sklearn.linear_model import Ridge

reg = Ridge(alpha=.1)
```

```python
reg.fit(train[predictors], train["Share"])
```

```python
predictions = reg.predict(test[predictors])
```

```python
predictions = pd.DataFrame(predictions, columns=["Predictions"], index=test.index)
```

```python
combination = pd.concat([test["Player"], predictions], axis=1)
```

```python
combination
```

```python
combination.sort_values("Predictions", ascending=False).head(10)
```

```python
combination = combination.sort_values("Predictions", ascending=False)
combination["Prediction_Rk"] = list(range(1,combination.shape[0]+1))
```

```python
combination.head(10)
```

```python
stat_ratios = stats[["PTS", "AST", "STL", "BLK", "3P", "Year"]].groupby("Year").apply(lambda x: x/x.mean())
```

```python
stats[["PTS_R", "AST_R", "STL_R", "BLK_R", "3P_R"]] = stat_ratios[["PTS", "AST", "STL", "BLK", "3P"]]
```
