```python
del players["Unnamed: 0"]
del players["Rk"]
```

```python
players["Player"] = players["Player"].str.replace("*", "", regex=False)
```

```python
def single_data(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row
    
players = players.groupby(["Player", "Year"]).apply(single_data)
```
